# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
driver.get(url)
```

```python
rank=[]  
name=[]             
artist=[]           
upload_d=[]         
views=[]            
z=0
```

```python
# Extracting all rank tags

rank_tags = driver.find_elements(By.XPATH, "//div[@class='mw-parser-output']/table[2]/tbody/tr/td[1]")

rank_tags
```

```python
#Now the text of the rank is inside the tag extracted above

for i in rank_tags:
    r = i.text
    rank.append(r)
    
rank
```

```python
#Lets extract all the tags having the name
name_tags = driver.find_elements(By.XPATH, "//div[@class='mw-parser-output']/table[2]/tbody/tr/td[2]")
```

```python
#Now the text of the name is inside the tag extracted above
for i in name_tags:
    l = len(i.text.split())
    n = i.text.split()[0 : l-1]
    n = ' '.join(n)
    n1 = i.text.split()[l-1]
    l1 = len(n1)
    for j in range(0, l1):
        if n1[j] == "[":
            z = j
    n1 = n1[0 : z]
    n = n+" "+n1
    name.append(n.replace('"',''))
    
name
```

```python
#Lets extract all the tags having the artist
artist_tags = driver.find_elements(By.XPATH, "//div[@class='mw-parser-output']/table[2]/tbody/tr/td[3]")
```

```python
#Now the text of the artist is inside the tag extracted above

for i in artist_tags:
    a = i.text
    artist.append(a)
    
artist
```

```python
#Lets extract all the tags having the upload date
upload_tags = driver.find_elements(By.XPATH, "//div[@class='mw-parser-output']/table[2]/tbody/tr/td[5]")
```

```python
# now the text of upload tags is extracted in above tag.

for i in upload_tags:
    u = i.text
    upload_d.append(u)
    
upload_d
```

```python
#Lets extract all the tags having the views
views_tags=driver.find_elements(By.XPATH, "//div[@class='mw-parser-output']/table[2]/tbody/tr/td[4]")

views_tags
```

```python
# Now the extracted text of views is in above tag.

for i in views_tags:
    v = i.text
    views.append(v)
    
views
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

df1 = pd.DataFrame({'Rank' : rank, 'Name' : name, 'Artist/Uploader' : artist, 'Upload Date' : upload_d, 'Views (billions)' : views}).set_index('Rank')

df1
```

```python
#getting the desired webpage.
url='https://github.com/'
driver.get(url)
time.sleep(2)
```

```python
#Getting explore button and clicking on it
explore = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()

#Selecting trending option
trend_url = driver.find_element(By.XPATH,"/html/body/div[4]/main/div[1]/nav/div/a[3]")
urls = trend_url.get_attribute("href")
driver.get(urls)
```

```python
#Creating empty list
#Creating empty list
URLs = []
repository_title = []
Description = []
Contributors = []
Language = []
lang = []

#Fetching urls for each repository
repository = driver.find_elements(By.XPATH,"//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    URLs.append(i.get_attribute("href"))

    
# Scraping Repository Title data
title = driver.find_elements(By.XPATH,"//h1[@class = 'h3 lh-condensed']")
for i in title:
    repository_title.append(i.text)

    
# Scraping data from all repository page
for i in URLs:
    driver.get(i)
    time.sleep(3)
    
    
    # Scraping Repository Description data
    try:
        desc = driver.find_element(By.XPATH,"//p[@class='f4 mt-3']")
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        
    # Scraping Contributors Count data
    try:
        contributor = driver.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        Contributors.append(contributor.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
        
        
    # Scraping Languages used data
    try:
        for i in driver.find_elements(By.XPATH,"//span[@class='color-text-primary text-bold mr-1']"):
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('-')
        
```

```python
#DATA FRAMEING
Github=pd.DataFrame({})
Github['Repository title'] = repository_title
Github['Repository description'] = Description
Github['Contributors count'] = Contributors
Github['Language used'] = Language
Github
```

```python
# Connecting to the webdriver

driver=webdriver.Chrome(r'chromedriver.exe')
```

```python
# Getting the webpage of mentioned url 
url = "https://archive.ics.uci.edu/"
driver.get(url)
```

```python
#Finding view all dataset button
view_dataset = driver.find_element(By.XPATH,"//tbody[1]//tr/td[2]/span[2]/a")    
page_url = view_dataset.get_attribute("href")
driver.get(page_url)
time.sleep(5)

#Getting page urls containing list of all datasets
all_lst = driver.find_element(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[1]/tbody/tr/td[2]/p/a")  
lst_url = all_lst.get_attribute("href")           
driver.get(lst_url)
time.sleep(5)

#fetching url for each dataset
data_url = driver.find_elements(By.XPATH,"//p[@class='normal']//b/a")    

urls = []     
for i in data_url:
    urls.append(i.get_attribute("href"))
```

```python
#Creating empty list
Data_name = []
Data_type = []
Task = []
Attr_type = []
Instances = []
n_attributes = []
Year = []
```

```python
#Scraping requiredinfo from urls
for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    #Scraping Dataset name
    try: 
        ds_name = driver.find_element(By.XPATH,"//span[@class='heading']")
        Data_name.append(ds_name.text)
    except NoSuchElementException:
        Data_name.append('-')
    time.sleep(1)
    
    
    #Scraping data type
    try:
        dtype = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr/td[2]")
        if dtype.text == "N/A": raise NoSuchElementException
        Data_type.append(dtype.text)
    except NoSuchElementException:
        Data_type.append('-')
    time.sleep(1)
    
    
    #Scraping Attribute type
    try:
        atype = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[2]")
        if atype.text == "N/A": raise NoSuchElementException
        Attr_type.append(atype.text)
    except NoSuchElementException:
        Attr_type.append('-')
    time.sleep(1)
    
    
    #scraping Task
    try:
        task = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[3]/td[2]")
        if task.text == "N/A": raise NoSuchElementException
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    time.sleep(1)
    
    
    #Scraping No of instances
    try:
        inst = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr/td[4]")
        if inst.text == "N/A": raise NoSuchElementException
        Instances.append(inst.text)
    except NoSuchElementException:
        Instances.append('-')
    time.sleep(1)
    
    
    #Scraping  No of attribute
    try:
        attr = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[4]")
        if attr.text == "N/A": raise NoSuchElementException
        n_attributes.append(attr.text)
    except NoSuchElementException:
        n_attributes.append('-')
    time.sleep(1)
    
    
    #scraping year
    try:
        year = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[6]")
        if year.text == "N/A": raise NoSuchElementException
        Year.append(year.text[:4])
    except NoSuchElementException:
        Year.append('-')
    time.sleep(1)
    
    #DATA FRAMEING
UCI=pd.DataFrame({})
UCI['Data Name'] = Data_name
UCI['Data Type'] = Data_type
UCI['Task'] = Task
UCI['Attribute type'] = Attr_type
UCI['No of instance'] = Instances
UCI['No of attributes'] = n_attributes
UCI['Year'] = Year

UCI
```
